```python
# Import Modules
import pandas as pd
from pathlib import Path 
```

```python
# Read CSV File into DF
grad_23 = Path("Resources/census_2023.csv")
df_23 = pd.read_csv(grad_23)

df_23
```

```python
# Rename Columns by taking out extra characters
df_23 = df_23.rename(columns=lambda col: col.replace('!!Total!!Estimate', ''))

df_23
```

```python
# Transpose DF and remove first row for a nice looking DF
df_23_t = df_23.transpose()
df_23_t.columns = df_23_t.iloc[0]
df_23_t = df_23_t[1:]
df_23_t.head()
```

```python
# Change index to State
df_23_t.index.name = "State name"
df_23_t.head()
```

```python
# Remove all commas from values and change all objects to integers
df_23_int = df_23_t.replace(",", "", regex = True)
df_23_int = df_23_int.astype(int)
df_23_int.dtypes
```

```python
# Add columns for totals and the graduation rate
df_23_int["2023 Total Population"] = 0
df_23_int["2023 Total High School Graduates"] = 0
df_23_int["2023 High School Grad Rate"] = 0

# Take out white space in column names to avoid errors
df_23_int.columns = df_23_int.columns.str.strip()

df_23_int.head()
```

```python
# Make calculations for empty columns
df_23_int["2023 Total Population"] = df_23_int["Population 18 to 24 years"] + df_23_int["Population 25 years and over"]
df_23_int["2023 Total High School Graduates"] = df_23_int["High school graduate (includes equivalency)"] + df_23_int["Some college or associate's degree"] + df_23_int["Bachelor's degree or higher"] + df_23_int["High school graduate or higher"]
df_23_int["2023 High School Grad Rate"] = round((df_23_int["2023 Total High School Graduates"] / df_23_int["2023 Total Population"])*100,2)
df_23_int.head()
```

Repeat the above steps for 2022 and 2021 data. To save space, I have added all steps into a single cell. 

```python
# Read CSV File into DF
grad_22 = Path("Resources/census_2022.csv")
df_22 = pd.read_csv(grad_22)

# Rename Columns by taking out extra characters
df_22 = df_22.rename(columns=lambda col: col.replace('!!Total!!Estimate', ''))

# Transpose DF and remove first row for a nice looking DF
df_22_t = df_22.transpose()
df_22_t.columns = df_22_t.iloc[0]
df_22_t = df_22_t[1:]

# Change index to State
df_22_t.index.name = "State name"

# Remove all commas from values and change all objects to integers
df_22_int = df_22_t.replace(",", "", regex = True)
df_22_int = df_22_int.astype(int)

# Add columns for totals and the graduation rate
df_22_int["2022 Total Population"] = 0
df_22_int["2022 Total High School Graduates"] = 0
df_22_int["2022 High School Grad Rate"] = 0

# Take out white space in column names to avoid errors
df_22_int.columns = df_22_int.columns.str.strip()

# Make calculations for empty columns
df_22_int["2022 Total Population"] = df_22_int["Population 18 to 24 years"] + df_22_int["Population 25 years and over"]
df_22_int["2022 Total High School Graduates"] = df_22_int["High school graduate (includes equivalency)"] + df_22_int["Some college or associate's degree"] + df_22_int["Bachelor's degree or higher"] + df_22_int["High school graduate or higher"]
df_22_int["2022 High School Grad Rate"] = round((df_22_int["2022 Total High School Graduates"] / df_22_int["2022 Total Population"])*100,2)


df_22_int.head()
```

```python
# Read CSV File into DF
grad_21 = Path("Resources/census_2021.csv")
df_21 = pd.read_csv(grad_21)

# Rename Columns by taking out extra characters
df_21 = df_21.rename(columns=lambda col: col.replace('!!Total!!Estimate', ''))

# Transpose DF and remove first row for a nice looking DF
df_21_t = df_21.transpose()
df_21_t.columns = df_21_t.iloc[0]
df_21_t = df_21_t[1:]

# Change index to State
df_21_t.index.name = "State name"

# Remove all commas from values and change all objects to integers
df_21_int = df_21_t.replace(",", "", regex = True)
df_21_int = df_21_int.astype(int)

# Add columns for totals and the graduation rate
df_21_int["2021 Total Population"] = 0
df_21_int["2021 Total High School Graduates"] = 0
df_21_int["2021 High School Grad Rate"] = 0

# Take out white space in column names to avoid errors
df_21_int.columns = df_21_int.columns.str.strip()

# Make calculations for empty columns
df_21_int["2021 Total Population"] = df_21_int["Population 18 to 24 years"] + df_21_int["Population 25 years and over"]
df_21_int["2021 Total High School Graduates"] = df_21_int["High school graduate (includes equivalency)"] + df_21_int["Some college or associate's degree"] + df_21_int["Bachelor's degree or higher"] + df_21_int["High school graduate or higher"]
df_21_int["2021 High School Grad Rate"] = round((df_21_int["2021 Total High School Graduates"] / df_21_int["2021 Total Population"])*100,2)


df_21_int.head()
```

```python
# Merge all three data frames using the "State" column and add columns of the graduation rates by year. 
grad_23_22 = pd.merge(df_23_int, df_22_int, on = ["State name"])

# Keep only Grad Rates 
keep_columns = ["2023 Total Population", "2023 High School Grad Rate", "2022 Total Population", "2022 High School Grad Rate"]
filtered_grad_23_22 = grad_23_22[keep_columns]

# Merge again
all_grad = pd.merge(filtered_grad_23_22, df_21_int, on = ["State name"])

# Keep only grad rates
keep_cols = ["2023 Total Population", "2023 High School Grad Rate", "2022 Total Population", "2022 High School Grad Rate", "2021 Total Population", "2021 High School Grad Rate"]
filtered_all_grad = all_grad[keep_cols]
filtered_all_grad = filtered_all_grad.astype(float)

# Calculate 3-year average
filtered_all_grad["Average High School Graduation Rate (2021-2023)"] = round(filtered_all_grad[["2021 High School Grad Rate", "2022 High School Grad Rate", "2021 High School Grad Rate"]].mean(axis=1), 2)
filtered_all_grad["Average Population (2021-2023)"] = filtered_all_grad[["2023 Total Population", "2022 Total Population", "2021 Total Population"]].mean(axis=1).astype("int")
# Export to CSV for group to use in project
filtered_all_grad.to_csv("Resources/Average_HS_Grad_Rates.csv", index = True, header = True)

filtered_all_grad.head()
```
