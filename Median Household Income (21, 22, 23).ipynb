```python
# Import Modules
import pandas as pd
from pathlib import Path 
```

```python
# Read CSV File for 2023
census_data_23 = Path("Resources/census_data_2023.csv")
census_df_23 = pd.read_csv(census_data_23)

# Display the DataFrame
census_df_23.head()

# Keep only columns that contain "Median income (dollars)!!Estimate" in their name
filtered_columns_23 = [col for col in census_df_23.columns if "Median income (dollars)!!Estimate" in col or "Label (Grouping)" in col]

# Create a new DataFrame with only the filtered columns
census_df_filtered_23 = census_df_23[filtered_columns_23]

# Check the filtered DataFrame
census_df_filtered_23.head()

# Transpose DF
census_df_filtered_t_23 = census_df_filtered_23.transpose()
census_df_filtered_t_23.columns = census_df_filtered_t_23.iloc[0]
census_df_filtered_t_23.head()

# Rename Columns by taking out extra characters
census_renamed_23 = census_df_filtered_23.rename(columns=lambda col: col.replace('!!', ''))
census_renamed_23

#transpose
census_t_23 = census_renamed_23.transpose()
census_t_23.head()
```

```python
census_t_23[1]
```

```python
census_23_households = census_t_23[1].reset_index()
```

```python
census_23_households.columns = census_23_households.iloc[0]
census_23_households = census_23_households[1:]
```

```python
census_23_households
```
